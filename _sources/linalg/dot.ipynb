```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plane(lim=[-5, 5], xlim=None, ylim=None):
    if lim is not None:
        xlim = ylim = lim
    plt.plot(xlim, [0, 0], color="black")
    plt.plot([0, 0], ylim, color="black")
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.xticks(np.arange(xlim[0], xlim[1]+1))
    plt.yticks(np.arange(ylim[0], ylim[1]+1))
    plt.grid()
```

```python
x = np.array([1, 1])

plt.figure(figsize=[6, 6])
plt.scatter(x[0], x[1])
plane()
None
```

```python
A = np.array([[1, 2], [-3, 3]])
y = A @ x
print(y)

plt.figure(figsize=[6, 6])
plt.scatter(x[0], x[1])
plt.scatter(y[0], y[1])
plt.arrow(*x, *(y-x), head_width=0.2, head_length=0.5, length_includes_head=True)
plane()
None
```

```python
x1 = np.repeat(np.linspace(-1, 1, 5), 5)
x2 = np.tile(np.linspace(-1, 1, 5), 5)
X = np.vstack([x1, x2])

A = np.array([[4, 3], [3, 4]]) / 3
Y = A @ X

plt.figure(figsize=[6, 6])
plt.scatter(X[0], X[1])
plt.scatter(Y[0], Y[1])
for i in range(25):
    # if i % 10 in (0, 5, 9):
    x = X[:, i]
    y = Y[:, i]
    plt.arrow(*x, *(y-x), head_width=0.2, head_length=0.5, length_includes_head=True)
plane()
None
```

```python
A = np.array([[4, 3], [3, 4]]) / 3
E = np.eye(2)
Z = np.random.multivariate_normal([0, 0], E @ A, size=400).T

plt.figure(figsize=[5, 5])
plt.scatter(Z[0], Z[1])
plane()
# plt.grid()
# plt.set(xlim=[-5, 5], ylim=[-5, 5])
```
